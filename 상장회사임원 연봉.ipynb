```python
import sys
import json
from urllib.request import urlopen
import pprint
from bs4 import BeautifulSoup
import re

baseURL1="http://dart.fss.or.kr/api/search.json?auth=0ab25f6f787f8da6b917a26f7a911ac18a836413&"
baseURL2="&end_dt=20190601&start_dt=20190101&fin_rpt=Y&dsp_tp=A&bsn_tp=A001&page_no=2&page_set=100&series=asc"

file=open("codes.csv", 'r')
for line in file:
    crp=line[:-1] # last is nl, which is removed
    response=urlopen(baseURL1+"crp_cd="+crp+baseURL2)
    data=response.read().decode('utf8')
    pp=pprint.PrettyPrinter(indent=1)
    pp.pprint(json.loads(data))
    decoded=json.loads(data)
    
    if decoded['err_code'] != '000':
        print('read error', baseURL1+baseURL2)
        sys.exit(1)
    for rpt in decoded['list']:
        if not "사업보고서" in rpt['rpt_nm']:
            continue
        if not "2018.12" in rpt['rpt_nm']:
            continue
        url="http://dart.fss.or.kr/dsaf001/main.do?rcpNo="+rpt['rcp_no']
        resp=urlopen(url)
        htm=resp.read().decode('utf8')
        print(htm)
        bs=BeautifulSoup(htm, "html.parser")
        for x in bs.findAll("a", onclick=True):
            if re.match('openPdfDownload', x['onclick']):
                args=x['onclick'].split()
                pattern=re.compile("openpdfdownload\(",re.IGNORECASE)
                rcpno=pattern.sub("",args[0])
                rcpno=rcpno.replace(",","")[1:-1] #remove first and last '
                dcmno=args[1].replace(");","")[1:-1] #remove first and last '
                pdfurl="http://dart.fss.or.kr/pdf/download/pdf.do?rcp_no="+rcpno+"&dcm_no="+dcmno
                print(crp, pdfurl)
                # save as pdf
                p=urlopen(pdfurl)
                with open(crp+"_"+dcmno+".pdf",'bw') as f:
                    f.write(p.read())
                f.close()
```
